## Пропущенные значения

Пропуски возникают по множеству причин:

- пользователь не заполнил поле
- данные потерялись при выгрузке
- значение не применимо к объекту

В Pandas пропуски обычно обозначаются как NaN.

### Поиск пропусков

```python
df.isna().sum()
```

У нас нет пропусков, но, в случае наличия, с ними надо поработать – исключить из данных или заполнить.

К заполнению пропуском существует несколько подходов. Выбор зависит от задачи. Ниже фрагмент кода показывает, как можно было бы заполнить пропуск по возрасту "средним" по имеющимся данным.

- различные варианты написания.

Одна и та же категория может быть записана по-разному:

- Москва
- москва
- МОСКВА
- Мск

Перед анализом такие значения нужно привести к единому виду.

Ниже вы увидите несколько примеров для масовой перекодировки значений.

```python
df["city"] = df["city"].replace(
    "мск",
    "москва"
)
```

```python
df["city"].value_counts()
```

```python
mapping = {
    "мск": "москва",
    "Москва": "москва",
    "MOSCOW": "москва",
    "спб": "санкт-петербург",
    "Питер": "санкт-петербург"
}

df["city"] = df["city"].replace(mapping)
```

```python
df["gender"] = df["gender"].replace({
    "м": "мужской",
    "муж": "мужской",
    "ж": "женский",
    "жен": "женский"
})
```

# Работа с датами

Дата часто загружается как обычный текст, а для анализа её нужно преобразовать в специальный формат.


## Стандартный формат даты

Для анализа удобнее всего приводить даты к формату:

$$YYYY-MM-DD$$

Например, 2024-03-15. Такой формат хорошо сортируется, однозначно читается и корректно преобразуется в datetime.

### Преобразование даты

```python
df["date"] = pd.to_datetime(df["date"])
```

### Получение года

```python
df["year"] = df["date"].dt.year
```

### Получение месяца

```python
df["month"] = df["date"].dt.month
```

## Если нужна детальная работа с датами

Когда важны конкретные дни, интервалы, длительность, сезонность или динамика по дням, дату лучше хранить как datetime.

```python
df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
df["day"] = df["date"].dt.day
df["weekday"] = df["date"].dt.day_name()
```

## Если анализ идёт по годам

Если день и месяц не важны, можно создать отдельную переменную year.

```python
df["year"] = df["date"].dt.year
```

## Если анализ идёт по месяцам

Для анализа по связке год + месяц лучше использовать период.

```python
df["year_month"] = df["date"].dt.to_period("M")
```

Получится формат 
2024-03
2024-04
2024-05

## Если нужен квартал

```python
df["quarter"] = df["date"].dt.to_period("Q")
```

# Создание новых признаков

## Зачем создавать новые переменные?

Новые переменные помогают сделать данные более удобными для анализа.

В примерах ниже новый признак создаётся в ходе математической операции на основе двух уже существующих признаков, как метапризнак или в ходе условной перекодировки.

- Разница между двумя столбцами

```python
df["profit"] = df["revenue"] - df["costs"]
```

- Отношение

```python
df["income_per_person"] = (df["income"] / df["family_size"])
```

```python
df["conversion"] = df["orders"] / df["visits"]
```

- Процент

```python
df["discount_share"] = df["discount"] / df["price"] * 100
```

- Логарифм

```python
df["log_income"] = np.log(df["income"])
```

- Длина текста

```python
df["comment_length"] = df["comment"].str.len()
```

- Категория на основе условия

```python
df["age_group"] = np.where(
    df["age"] >= 18,
    "adult",
    "minor"
)
```

- Создание переменной с несколькими условиями

```python
conditions = [
    df["age"] < 18,
    (df["age"] >= 18) & (df["age"] < 35),
    (df["age"] >= 35) & (df["age"] < 60),
    df["age"] >= 60
]

values = [
    "до 18",
    "18–34",
    "35–59",
    "60+"
]

df["age_group"] = np.select(
    conditions,
    values,
    default="нет данных"
)
```

- Создание признака из даты

```python
df["is_weekend"] = df["date"].dt.weekday >= 5
```

- Создание признака из текста

```python
df["has_delivery"] = df["comment"].str.contains(
    "доставка",
    case=False,
    na=False
)
```

- Создание признака из нескольких столбцов

```python
df["full_name"] = (
    df["last_name"] + " " + df["first_name"]
)
```
