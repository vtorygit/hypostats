# Переменные

## Что такое переменная

Во время анализа данных нам постоянно приходится сохранять результаты вычислений.

Для этого используются переменные.

Слева находится имя переменной.

Справа — значение, которое в ней хранится.

```python
age = 25
```

Переменные позволяют нам не писать значение каждый раз, а положить его в специальную коробочку и использовать столько раз, сколько потребуется. Если нужно изменить значение, менять его нужно только в одном месте – там, где значение переменной задаётся. 

```python
age = 25

print(age)
```

Важно следить за тем, какие имена вы выбираете для переменых – не используйте служебные названия (те, что уже есть в python и других языках программирования, скорее всего среда подсветит их при попытке ввода) и не повторяйтесь в названиях в рамках одного кода.

Представим, что мы записываем оценки 3 девушек, двух из которых зовут Аня. Одна Аня поставила оценку 10, а другая 5. Считая суммарную оценку, мы бы сложили 10 + 5 + 5 = 20, но в коде ниже вы увидите, что при вводе оценки второй Ани изменилось значение самой переменной Ann – там было значение 10, а стало значение 5, оно же два раза вошло в расчёт.

```python
Ann = 10
Gyll = 5
Ann = 5
print(Ann + Gyll + Ann)
```

В работе с данными вы часто будете переиспользовать код и в нём часто зашиты разные переменные. Если никак этим не управлять, то, если запустить тест два раза на разных данных, в переменных сохранятся результаты только последнего запуска, то есть первые результаты будут перезаписаны. Учитывайте это!

# Циклы

Циклы позволяют повторять одно и то же действие несколько раз.

В Python чаще всего используются два вида циклов:

- for;
- while.

## Цикл for

Цикл `for` используется, когда мы заранее знаем, по каким элементам нужно пройтись.

Например, у нас есть список оценок.

```python
scores = [78, 82, 91, 65, 87]

for score in scores:
    print(score)
```

На каждом шаге цикла переменная score принимает очередное значение из списка.

## Пример с вычислением

```python
scores = [78, 82, 91, 65, 87]

for score in scores:
    print(score + 5)
```

## Цикл while

Цикл `while` используется, когда действие нужно повторять до тех пор, пока выполняется условие.

```python
count = 1

while count <= 5:
    print(count)
    count = count + 1
```

Сначала Python проверяет условие. Если условие истинно, выполняется тело цикла. После этого условие проверяется снова.

## Разница между for и while

`for` обычно используется, когда есть набор элементов:

- список
- массив
- строки таблицы
- названия столбцов

`while` используется, когда заранее неизвестно, сколько раз нужно повторить действие.

```python
# for: пройти по списку городов
cities = ["москва", "казань", "самара"]

for city in cities:
    print(city)
```

```python
# while: повторять, пока значение меньше порога
value = 1

while value < 100:
    value = value * 2
    print(value)
```

## Циклы в анализе данных

В анализе данных циклы часто используются, чтобы применить одно действие к нескольким столбцам.

```python
import pandas as pd

df = pd.DataFrame({
    "age": [21, 25, 19, 30, 27],
    "income": [50000, 65000, 42000, 80000, 70000],
    "score": [78, 85, 72, 91, 88]
})

columns = ["age", "income", "score"]

for column in columns:
    print(column)
    print(df[column].mean())
```

При этом в Pandas многие операции можно делать без циклов.

Например, среднее по нескольким столбцам можно посчитать так:

```python
df[["age", "income", "score"]].mean()
```

Циклы полезно понимать, но в анализе данных часто лучше использовать готовые методы Pandas.
